```python
import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt
```

```python
f_x = lambda x: np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
```

```python
poly_rank = 1
A_1 = np.array([[1 ** n for n in range(0, poly_rank + 1)], [15 ** n for n in range(0, poly_rank + 1)]])
b_1 = np.array([f_x(1), f_x(15)])
x_1 = scipy.linalg.solve(A_1, b_1)
```

```python
poly_rank = 2
A_2 = np.array([[1 ** n for n in range(0, poly_rank + 1)], [8 ** n for n in range(0, poly_rank + 1)], [15 ** n for n in range(0, poly_rank + 1)]])
b_2 = np.array([f_x(1), f_x(8), f_x(15)])
x_2 = scipy.linalg.solve(A_2, b_2)
```

```python
poly_rank = 3
A_3 = np.array([[1 ** n for n in range(0, poly_rank + 1)], [4 ** n for n in range(0, poly_rank + 1)], [10 ** n for n in range(0, poly_rank + 1)], [15 ** n for n in range(0, poly_rank + 1)]])
b_3 = np.array([f_x(1), f_x(4), f_x(10), f_x(15)])
x_3 = scipy.linalg.solve(A_3, b_3)
```

```python
print(x_3)
```

```python
print(list(map(lambda x: x.round(2), x_3)))
```
